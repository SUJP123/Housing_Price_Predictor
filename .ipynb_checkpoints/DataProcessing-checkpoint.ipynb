```python
import pandas as pd

fed_files = ["datasets/MORTGAGE30US2024.csv", "datasets/RRVRUSQ156N2024.csv","datasets/CPIAUCSL2024.csv"]

dfs = [pd.read_csv(f, parse_dates=True, index_col=0) for f in fed_files]
```

```python
dfs[1]
```

```python
fed_data = pd.concat(dfs, axis=1)
fed_data = fed_data.ffill()
```

```python
fed_data.tail(20)
```

```python
zillow_files = ["datasets/Metro_median_sale_price_uc_sfrcondo_sm_sa_week2024.csv", "datasets/Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month2024.csv"] 

dfs = [pd.read_csv(f) for f in zillow_files]
```

```python
dfs[0]
```

```python
Regions = dfs[0]['RegionName']
Regions[2]
```

```python
from datetime import timedelta

def generateZillowData(index):
    dfs1 = [pd.DataFrame(df.iloc[index,5:]) for df in dfs]
    region = Regions[index]
    for df in dfs1:
        df.index = pd.to_datetime(df.index)
        df["month"] = df.index.to_period("M")
        df["region"] = region
    price = dfs1[0].merge(dfs1[1], on=["month","region"])
    price.index = dfs1[0].index
    del price["month"]
    price.columns = ["price", "region", "value"]
    return price

def mergeFedAndZillow(fed, zillow):
    fed = fed.dropna()
    fed.index = fed.index + timedelta(days=2)
    price_data = fed.merge(zillow, left_index=True, right_index=True)
    price_data.columns = ["interest", "vacancy", "cpi", "price","region","value"]
    price_data["adj_price"] = price_data["price"] / price_data["cpi"] * 100
    price_data["adj_value"] = price_data["value"] / price_data["cpi"] * 100
    price_data["next_quarter"] = price_data["adj_price"].shift(-13)
    return price_data
```

```python
#Test Data for LA
x = generateZillowData(2)
price_data = mergeFedAndZillow(fed_data, x)
price_data.dropna(inplace=True)
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)
price_data.tail(50)
```

```python
price_data["change"].value_counts()
```

```python
predictors = ["interest", "vacancy", "adj_price", "adj_value"]
target = "change"
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
START = 260
STEP = 52

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])

    return preds

def backtest(data, predictors, target, dataset):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = dataset.iloc[:i]
        test = dataset.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target, price_data)
```

```python
accuracy
```

```python
preds
```

```python
pred_match = (preds == price_data[target].iloc[START:])
```

```python
x = price_data.copy()
del x["region"]
yearly = x.rolling(52, min_periods=1).mean()
yearly_ratios = [p+"_year" for p in predictors]
x[yearly_ratios] = x[predictors] / yearly[predictors]
x
```

```python
preds, accuracy = backtest(x, predictors + yearly_ratios, target, x)
```

```python
accuracy
```

```python
pred_match = (preds == price_data[target].iloc[START:])
pred_match[pred_match == True] = 1
pred_match[pred_match == False] = 0
pred_match[pred_match == 1] = "green"
pred_match[pred_match == 0] = "red"
```

```python
import matplotlib.pyplot as plt

plot_data = price_data.iloc[START:].copy()
plot_data.reset_index().plot.scatter(x="index", y="adj_price", color=pred_match)
```

```python
import os

def save_dataframe_to_csv(df, directory, file_name):
    """
    Save the DataFrame as a CSV file in the specified directory.
    If a file with the same name exists, it will be replaced.
    
    :param df: DataFrame to save
    :param directory: Directory where the file will be saved
    :param file_name: Name of the CSV file
    """
    # Create the directory if it doesn't exist
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Construct the full file path
    file_path = os.path.join(directory, file_name)
    
    # Save the DataFrame to a CSV file, replacing any existing file with the same name
    df.to_csv(file_path, index_label='date')
    print(f'DataFrame saved to {file_path}')
```

```python
direct = 'processed_data/'

for i in range(len(Regions)):
    data = mergeFedAndZillow(fed_data, generateZillowData(i))
    data = data.drop(['region'], axis=1)
    filename = Regions[i].replace(" ","").replace(",","") + ".csv"
    save_dataframe_to_csv(data, direct, filename)
```
